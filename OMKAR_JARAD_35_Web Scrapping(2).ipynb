```python
import requests
from bs4 import BeautifulSoup
```

```python
url=f"https://www.amazon.in/product-reviews/B01BOWU7YG/ref=cm_cr_arp_d_paging_btm_next_2?pageNumber=1"
headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"}
```

```python
response=requests.get(url, headers)
print(response)
```

```python
response=response.content 
```

```python
soup=BeautifulSoup(response, 'html.parser')
soup
```

```python
data=soup.find('div', class_='a-section a-spacing-none review-views celwidget')
revives=data.find_all('div',class_='a-section review aok-relative') 
```

```python
product_revives_data=[]
for revive in revives:
    customer_name=revive.find('span').text 
    rating=revive.find('span', class_='a-icon-alt').text
    revive_tag=revive.find(class_='a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold')
    revive_tag_text=revive_tag.find('span').text 
    product_revive=revive.find('span',class_='a-size-base a-color-secondary review-date').text
    comment=revive.find('span', class_='a-size-base review-text review-text-content').text
    verified=revive.find('span',class_='a-size-mini a-color-state a-text-bold').text
    product_revives_data.append([customer_name,revive_tag_text,rating,product_revive,comment,verified])
print(product_revives_data)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(product_revives_data, columns=["customer_name","revive_tag_text","rating","product_revive","comment","verified"])
```

```python
df.to_csv('revives.csv')
```
